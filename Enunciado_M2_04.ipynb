**3) Realizar un programa que pida al usuario 3 números que pueden ser flotantes (no es necesario realizar bucles aún, podemos repetir el código), estos números se deberán introducir en una lista. Cuando se haya finalizado la introducción de los datos, se mostrará el sumatorio de toda la lista. Guardar el sumatorio en una variable que se llame "sumatorio"**

*Ayuda: Existe una función llamada sum(lista) ¡Pruébalo!*

```python
# Ejercicio 3: Pedir al usuario 3 números (pueden ser flotantes) y guardarlos en una lista
numeros = []
# Repetimos el input 3 veces (sin usar bucles avanzados en las instrucciones, pero aquí usamos for para concisión)
for i in range(1, 4):
    entrada = input(f'Introduce el número {i}: ')
    try:
        n = float(entrada)
    except ValueError:
        print('Entrada inválida, se guardará 0.0')
        n = 0.0
    numeros.append(n)
# Calculamos el sumatorio y lo guardamos en la variable 'sumatorio'
sumatorio = sum(numeros)
print('Números introducidos:', numeros)
print('Sumatorio:', sumatorio)
```
